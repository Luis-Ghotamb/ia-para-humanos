# üìà Regress√£o Linear - IA para Humanos

> Vamos prever pre√ßos de casas com base em seu tamanho. Um exemplo cl√°ssico para entender como IA pode aprender padr√µes simples.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
# Dados: tamanho da casa (m¬≤) vs. pre√ßo (R$ mil)
X = np.array([50, 60, 70, 80, 90, 100, 110, 120]).reshape(-1, 1)
y = np.array([150, 180, 200, 210, 240, 250, 270, 290])
```

```python
modelo = LinearRegression()
modelo.fit(X, y)

print(f"Coeficiente (m): {modelo.coef_[0]:.2f}")
print(f"Intercepto (b): {modelo.intercept_:.2f}")
```

```python
plt.scatter(X, y, color='blue', label='Dados reais')
plt.plot(X, modelo.predict(X), color='red', label='Regress√£o Linear')
plt.title('Pre√ßo x Tamanho da Casa')
plt.xlabel('Tamanho (m¬≤)')
plt.ylabel('Pre√ßo (R$ mil)')
plt.legend()
plt.grid(True)
plt.show()
```

## üß† Conclus√£o

A regress√£o linear encontra a melhor **reta** que passa pelos dados, minimizando os erros entre o valor previsto e o valor real.

O modelo aprendeu a f√≥rmula:  
`pre√ßo = m * tamanho + b`  
Onde `m` √© o coeficiente (inclina√ß√£o da reta) e `b` √© o intercepto (ponto onde a reta cruza o eixo Y).
